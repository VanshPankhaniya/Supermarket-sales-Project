# 2. On which date is the highest gross margin percentage

```python
plt.subplots(figsize=(18,10))
ax=sns.scatterplot(x = vp["Date"], y = vp["gross margin percentage"],color='r')
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.grid()
```

# 13. All country's supermarket sales with pie chart 

```python
x = np.array([74.22, 10.56, 62.57, 11.85, 91.3 , 40.73, 52.38, 38.54, 44.63,])
mylabels = ("India","Australia", "Canada", "China", "Cambodia", "Denmark", "Hong Kong", "Israel","Malaysia")
plt.pie(x , labels = mylabels , autopct = "%1.2f%%")
plt.legend(title = "All Countrys")
plt.title("All country's supermarket sales")
plt.show()
```

# 14. Which is the hightest rating of supermarket 

```python
fig , ax = plt.subplots(figsize = (25,36))
sns.countplot(x = "Rating", data = vp , ax = ax)
for i in ax.containers:
    ax.bar_label(i)
plt.title("Supermarket Sales")
```
